Visualization

Distribution of Dog Breeds by Popularity Ranking

```python
# Sort the data by 'popularity ranking'
df_sorted = df.sort_values(by='popularity ranking')

# Limit the number of breeds to top N (e.g., top 20)
top_n = 20
df_top_n = df_sorted.head(top_n)

# Create the plot
plt.figure(figsize=(12, 8))  # Adjust the figure size
sns.barplot(x='popularity ranking', y='Breed', data=df_top_n)

# Add titles and labels
plt.title('Top Dog Breeds by Popularity Ranking')
plt.xlabel('Popularity Ranking')
plt.ylabel('Breed')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

Longevity vs. Purchase Price

```python
# Clean the 'PURCHASE PRICE' column by removing '$' and ',' characters and converting to float
df['PURCHASE PRICE'] = df['PURCHASE PRICE'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Adjust the figure size
plt.figure(figsize=(10, 6))

# Create the scatter plot with transparency (alpha), adjusted marker size, and axis limits
sns.scatterplot(x='LONGEVITY(YEARS)', y='PURCHASE PRICE', data=df, alpha=0.6, s=100, color='b')

# Set the axis limits (optional, depending on your data)
plt.xlim(df['LONGEVITY(YEARS)'].min() - 1, df['LONGEVITY(YEARS)'].max() + 1)
plt.ylim(df['PURCHASE PRICE'].min() - 1000, df['PURCHASE PRICE'].max() + 1000)

# Add titles and labels
plt.title('Longevity vs Purchase Price', fontsize=16)
plt.xlabel('Longevity (Years)', fontsize=14)
plt.ylabel('Purchase Price ($)', fontsize=14)

# Show the plot
plt.show()
```

Intelligence vs. Longevity

```python
# Convert 'intelligence' column to categorical if not already
df['intelligence'] = df['intelligence'].astype('category')

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='intelligence', y='LONGEVITY(YEARS)', data=df)

# Add titles and labels
plt.title('Intelligence vs Longevity (Box Plot)', fontsize=16)
plt.xlabel('Intelligence', fontsize=14)
plt.ylabel('Longevity (Years)', fontsize=14)

# Show the plot
plt.show()
```

- In general, fair to above average intelligence results in very similar longevity from about 9 to 13 years of age.
- For dogs who fair in lowest intelligence, they also live just as long but also pass away much sooner at about 8 years of age. This could be due to other factors unrelated to intelligence!

Intelligence vs. Popularity

```python
sns.heatmap(df[['INTELLIGENCE RANK', 'popularity ranking']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Intelligence and Popularity')
plt.show()
```

Dogs that are considered more intelligent may be slightly more popular, but intelligence is not a major factor driving popularity in this dataset.

Breed Size vs. Purchase Price

```python
sns.boxplot(x='size', y='PURCHASE PRICE', data=df)
plt.title('Breed Size vs Purchase Price')
plt.show()
```

The results here show that every so slighly that larger dogs are a tad more expensive in about the $1200 range but small dogs are at about $900. This difference is more likely due to poplarity and breed quality. Breed quality is highly important and we must consider factors such as reputation of breeders, OFA results (eyes and hip health), champion lineage of parents, etc. 

Number of Genetic Ailments vs. Longevity

```python
sns.scatterplot(x='NUMBER OF GENETIC AILMENTS', y='LONGEVITY(YEARS)', data=df)
plt.title('Genetic Ailments vs Longevity')
plt.show()
```

Obviously, the fewer the genetic ailments end up living longer as their health is much better. There are some outliers here with many genetic ailments living well into 13 years of age. 

Lifetime Costs vs. Longevity

```python
df['Lifetime Cost'] = df['$LIFETIME COST'].replace({'\$': '', ',': ''}, regex=True).astype(float)
sns.scatterplot(x='LONGEVITY(YEARS)', y='Lifetime Cost', data=df)
plt.title('Lifetime Costs vs Longevity')
plt.show()
```

At roughly 10 years up until 16 years of age, there is a spread of lifetime cost ranging from as long as 16,000 up to 27,000. This analysis explains that the longer a dog lives, the more expensive it will cost as any living thing that ages requires medications due to breakdown of body and mind. It would be interesting to see which dog breeds are the healthiest.

Which dog breeds are the healthiest?

```python
# Filter breeds with no genetic ailments and no congenital ailments
healthy_breeds = df[(df['NUMBER OF GENETIC AILMENTS'] == 0) & (df['congential ailments'] == 'none')]

# Sort by longevity to find the healthiest (longest-living) breeds
healthy_breeds_sorted = healthy_breeds.sort_values(by='LONGEVITY(YEARS)', ascending=False)

# Display the top healthiest breeds
healthy_breeds_sorted[['Breed', 'NUMBER OF GENETIC AILMENTS', 'congential ailments', 'LONGEVITY(YEARS)']]
```
